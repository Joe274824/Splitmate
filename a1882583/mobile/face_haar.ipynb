```python
import cv2
import numpy as np
import os
```

```python
#需要用自己训练的标签和对应的人名填入完成这个字典
names = {1: 'ChenZhuo',
         2: 'bao',}
```

```python
recognizer = cv2.face.LBPHFaceRecognizer_create()
#下面这行需要填入你刚才输出的yml文件的绝对路径，例如/Users/xxx/Documents/ipynb/opencv_demo/trainer_4.yml
recognizer.read('')
def face_detect(img):
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) 
    #下面这行需要填入级联分类起的绝对路径，如/Users/xxx/Documents/ipynb/haarcascade/haarcascade_frontalface_alt2.xml
    face_detector = cv2.CascadeClassifier("")
    faces = face_detector.detectMultiScale(gray) 
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
        id,confidence = recognizer.predict(gray[y:y+h,x:x+w])
        cv2.putText(img, str(confidence), (x+w,y+h), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255,0), 3)
        if confidence <80 :
            name = names.get(id, "Unknown")
            cv2.putText(img, name, (x,y), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255,0), 3)
#     print(type(confidence))
        print('标签:',id,'置信评分:',confidence)
    cv2.namedWindow('result',cv2.WINDOW_NORMAL) 
    cv2.imshow('result',img)
```

```python
cap = cv2.VideoCapture(0)
cap.read()
while True:
    flag,frame = cap.read()
    if not flag:
        break 
    face_detect(frame)
    if ord('q') == cv2.waitKey(10):
        break 
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### 下面这段代码可以将你的代码所展现出来的视频保存到本地

```python
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
cap = cv2.VideoCapture('')#这里填写传入的视频的路径，如果需要打开摄像头，填写0即可
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
videoWrite = cv2.VideoWriter('', fourcc, 24, (width, height))#单引号之间要填写你视频保存的绝对路径
cap.read()
while True:
    flag,frame = cap.read()
    if not flag:
        break 
    face_detect(frame)
    videoWrite.write(frame)
    if ord('q') == cv2.waitKey(10):
        break 
cap.release()
videoWrite.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
